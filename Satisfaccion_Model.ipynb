```python
from sklearn.model_selection import train_test_split
from collections import Counter
from sklearn.preprocessing import StandardScaler
```

```python
features=df.columns
features=features[:-1]
```

```python
bin_satisfaccion = {"satisfied": 1, "neutral or dissatisfied": 0}

df['satisfaction'] = df["satisfaction"].apply(lambda x: bin_satisfaccion[x])
```

```python
y = df['satisfaction']
df = df.drop(columns=['satisfaction','Gender','Customer.Type','Type.of.Travel','Class'], axis=1)
scaler = StandardScaler().fit(df)

X = scaler.transform(df)
```

```python
X1, X_test, y1, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X1, y1, test_size=0.177, random_state=42)
nrow=X1.shape[0]
ncol_tot=X1.shape[1]
```

## Creacion de modelos

```python
from sklearn import tree
from sklearn.ensemble import (
    BaggingClassifier,
    RandomForestClassifier,
    GradientBoostingClassifier
)
from sklearn.model_selection import GridSearchCV
from sklearn.tree import plot_tree
from sklearn.metrics import  roc_auc_score, roc_curve, confusion_matrix, auc
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from sklearn import metrics
import kerastuner  as kt
from keras.initializers import GlorotNormal
from keras.optimizers import Adam
from keras.layers import  Dropout
import tensorflow as tf
from keras.initializers import RandomUniform
import keras

def plot_roc_curve(y_train_obs, y_train_pred,y_val_obs, y_val_pred,y_test_obs, y_test_pred):
    """
    plots the roc curve based of the probabilities
    """

    fpr, tpr, thresholds = roc_curve(y_train_obs, y_train_pred)
    plt.plot(fpr, tpr)

    fpr, tpr, thresholds = roc_curve(y_val_obs, y_val_pred)
    plt.plot(fpr, tpr)

    fpr, tpr, thresholds = roc_curve(y_test_obs, y_test_pred)
    plt.plot(fpr, tpr)

    plt.title('Curvas ROC por Conjunto de Datos')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend(["Train","Val","Test"])
    plt.grid(True)
    plt.show()

    print('\nAUC score Train:',round(roc_auc_score(y_train_obs, y_train_pred),2) )
    print('\nAUC score Val:',round(roc_auc_score(y_val_obs, y_val_pred),2) )
    print('\nAUC score Test:',round(roc_auc_score(y_test_obs, y_test_pred),2) )

# Función que mostrara el resumen para el entrenamiento y prueba
def summary_CM(actual, predicted):
    a=accuracy_score(actual, predicted)
    b=precision_score(actual, predicted)
    c= recall_score(actual, predicted)
    d=recall_score(actual, predicted, pos_label=0)
    e=f1_score(actual, predicted)
    return round(a,4),round(b,4),round(c,4),round(d,4),round(e,4)

def summary_train_val_test(y_train_obs, y_train_pred,y_val_obs, y_val_pred,y_test_obs, y_test_pred):
  train=summary_CM(y_train_obs,y_train_pred)
  val=summary_CM(y_val_obs,y_val_pred)
  test=summary_CM(y_test_obs,y_test_pred)
  df=pd.DataFrame([train,val,test],index=["Train","Val","Test"],columns=["Acc","Prec","Recall","Spec","F1"])
  return df

def pred_ajust(y_pred_p, par):
  y_pred=y_pred_p
  y_pred[y_pred>=par]=1
  y_pred[y_pred<par]=0
  y_pred=y_pred.astype(int)
  return y_pred

def evaluar_hiperparametro(hiper_par, yval_pred1_p, y_val):
  """
  Función que evalúa la precisión, la precisión y la puntuación F1 para diferentes valores de hiperparámetro.
  """

  resultados = []

  for i in hiper_par:
    yval_pred1 = pred_ajust(yval_pred1_p, i)
    acc = round(accuracy_score(y_val, yval_pred1), 4)
    prec = round(precision_score(y_val, yval_pred1), 4)
    f1 = round(f1_score(y_val, yval_pred1), 4)
    resultados.append([i, acc, prec, f1])

  return pd.DataFrame(resultados, columns=["Hiperparámetro", "Accuracy", "Precisión", "F1 Score"])

def build_model(hp):
    seed_value=3
    tf.random.set_seed(seed_value)
    hp_learning_rate = hp.Choice('learning_rate', values = [0.001, 0.01, 0.1])

    initializer = GlorotNormal(seed=seed_value)
    optimizer1 = Adam(learning_rate=hp_learning_rate)

    model = Sequential()
    model.add(Dense(units=hp.Int('layers_units_x', 50, 350, step=10), input_dim = ncol_tot,  kernel_initializer=initializer, activation='relu'))

    for i in range(hp.Int('layers_', 1, 6)):
        model.add(Dropout(rate=hp.Choice('dp_rate_'+ str(i), values = [0.25, 0.5, 0.75])))
        model.add(Dense(units=hp.Int('units_' + str(i), 50, 350, step=10),activation='relu',kernel_initializer=initializer))

    model.add(Dense(1, activation='sigmoid',kernel_initializer=initializer))

    model.compile(optimizer=optimizer1, loss='binary_crossentropy')
    return model
```

### Validacion cruzada

```python
from sklearn.model_selection import KFold

def validacion_cruzada(modelo):
  n_splits = 10
  val_Acc = []
  val_Prec = []
  val_F1 = []

  kf = KFold(n_splits=n_splits,shuffle=True,random_state = 3)

  mdls=["model_1","model_2","model_3","model_4","model_5", "model_6","model_7","model_8","model_9","model_10"]

  i=0

  for train_index, val_index in kf.split(X1):
    locals()[mdls[i]]=modelo
    X_train, X_val, y_train, y_val = train_test_split(X1, y1, test_size=0.177, random_state=42)

    nrow=X_train.shape[0]

    # Fit the model
    locals()[mdls[i]].fit(X_train, y_train, validation_data=(X_val, y_val),epochs=500, batch_size=nrow,verbose=0)
    print(mdls[i])
    yval_pred_p=locals()[mdls[i]].predict(X_val,verbose=0)
    yval_pred = pred_ajust(yval_pred_p, 0.5)
    metri=summary_CM(y_val,yval_pred)
    val_Acc.append(metri[0])
    val_Prec.append(metri[1])
    val_F1.append(metri[4])
    i=i+1

  print("Accuracy obtenido por cada partición en la validación ",val_Acc)
  print("Precision obtenido por cada partición en la validación ",val_Prec)
  print("F1-score obtenido por cada partición en la validación ",val_F1)

  print("Media del Accuracy de validación",np.mean(val_Acc))
  print("Media de la Precision de validación",np.mean(val_Prec))
  print("Media del F1-Score de validación",np.mean(val_F1))
```
